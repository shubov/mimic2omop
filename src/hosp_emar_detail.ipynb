```python
import pandas as pd
import numpy as np
```

```python
# read csv file

df = pd.read_csv('/data/mimic-iv-2.2/shubov_thesis/hosp/emar_detail.csv.gz', compression='gzip')
```

```python
df.head()
```

```python
df = df.replace('___', np.nan)
```

```python
print(df.dtypes)
```

```python
df['infusion_rate'] = df['infusion_rate'].astype(str)
```

```python
df['prior_infusion_rate'] = df['prior_infusion_rate'].astype(str)
```

```python
df['parent_field_ordinal'] = df['parent_field_ordinal'].astype(float)
```

```python
df['product_amount_given'] = df['product_amount_given'].apply(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
df['product_amount_given'].replace(['null', 0, '', None, 'None', 'NaN'], np.nan, inplace=True)
```

```python
df['product_amount_given'] = df['product_amount_given'].astype(float)
```

```python
df['pharmacy_id'] = df['pharmacy_id'].astype(pd.Int64Dtype())
```

```python
df['dose_given'] = df['dose_given'].astype(str)
```

```python
df.head()
```

```python
# compress the dataframe to gzip and save it as a csv file
df.to_csv('/data/mimic-iv-2.2/shubov_thesis/hosp/emar_detail.csv.gz', compression='gzip', index=False)
```
